## Linear regression algorithm from scratch

```python
#importing required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
X = np.arange(0,10)
y = X*2 + 0.5
X,y
```

```python
# we will assume thetas as a, b where y = ax + b 
#initial values of thetas is 0
a = 0 
b = 0 
```

```python

def computeCost ( X, y, a, b): 
    h = X.dot(a) + b 
    J = (1/(2*len(X)))*np.sum((h - y)**2)
    return J

def gradient(X, y, a , b, alpha, num_iter):
    J_values =[]
    a_values =[]
    b_values =[]
    for i in range(num_iter):
        #the h(theta) function
        h = X.dot(a) + b 

        #finding the descent after every iteration 
        update_a = alpha*((1/len(X))* ( np.sum((h.T - y).dot(X))))
        update_b = alpha*((1/len(X))* ( np.sum(h.T - y)))

        #simultaneously updating both the parameters
        a = a - update_a
        b = b - update_b
        
        a_values.append(a)
        b_values.append(b)
        J_values.append(computeCost(X,y,a,b))

    plt.plot(J_values)
    plt.xlabel('Iterations')
    plt.xticks(np.arange(0, num_iter + 5,step =5))
    plt.ylabel('Cost function')
    plt.title('Cost function vs Iterations')
    return J_values, a_values, b_values


final_cost , a_final, b_final = gradient(X,y,a,b,0.01,20)

print("Y = X*" + str(round(a_final[-1],2)) + " +" + str(round(b_final[-1],2)) + " with a cost minimised to " + str(final_cost[-1]))
```
