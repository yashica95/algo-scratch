```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
from logisticRegression import LogisticRegressionModal
```

```python
X, y  = load_iris(return_X_y=True)
```

```python
#keeping only two classes - 0 and 1 
X_train = X[:100, :]
y_train = y[:100]
```

```python
model = LogisticRegressionModal(print_details=True)
fit = model.fit(X_train, y_train, alpha=0.1, threshold=0.5, num_epochs=100)
```

```python
#predict the first 10 samples
model.predict(X_train[:10,:])
```
